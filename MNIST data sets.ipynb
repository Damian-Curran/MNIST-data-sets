```python
import gzip
import numpy as np
import PIL.Image as pilImage

#Function to read in images for both Train and Test sets
def imagess(filepath):
    with gzip.open(filepath, 'rb') as f:     
        
        #Save variables that are not image bytes
        magicNum = (int.from_bytes(f.read(4), "big"))
        noImages = (int.from_bytes(f.read(4), "big"))
        noRow = (int.from_bytes(f.read(4), "big"))
        noCol = (int.from_bytes(f.read(4), "big"))
                
        images = []
        
        for i in range(noImages):
            row = []
            for k in range(noRow):
                col = []
                for j in range(noCol):
                    col.append(int.from_bytes(f.read(1), "big"))
                row.append(col)
            images.append(row)     
            
        return images

#Function to read in labels for both Train and Test sets
def labelss(filepath):
        with gzip.open(filepath, 'rb') as f:
            labels = []
            
            #Saves variables that arent bytes for labels
            magicNum = (int.from_bytes(f.read(4), "big"))
            noLabels = (int.from_bytes(f.read(4), "big"))
            
            #for loop for range, 60,000/10,000
            for i in range(noLabels):
                #Reads byte at a time and appends to list
                labels.append((int.from_bytes(f.read(1), "big")))
        
        return labels

#Function to print out images in cells
def printMe(image):
count = 0

for rows in image:
    for col in rows:
        count += 1
        if(col <= 128):
            print(".", end = " ")
        else:
            print("#", end = " ")
        
        #forms new line after 28 bytes printed
        if(count == 28):
            print()
            count = 0

#Function to save pixels into .pngs
def save_image(image, tag, index, label):
    #Location to save images
    target = "Pictures/%s-%05d-%d.png"
    
    #Saves list value into np.array, into pixels
    pixels = np.array(image)
    img = pilImage.fromarray(pixels.astype('uint8'))
    #saves to target with ids
    img.save(target % (tag, index, label))        


#Save Function result into value
train_Images = imagess("C:\\users\\sgtcu\\Desktop\\ProblemSheet2\\trainingImages.gz")
train_Labels = labelss("C:\\users\\sgtcu\\Desktop\\ProblemSheet2\\trainingLabels.gz")

#Save function result into value
test_Images = imagess("C:\\users\\sgtcu\\Desktop\\ProblemSheet2\\testImages.gz")
test_Labels = labelss("C:\\users\\sgtcu\\Desktop\\ProblemSheet2\\testLabels.gz")

#Call function to print out specified image to cell
printMe(train_Images[2])

#For loop length of train_images(60,000)
for i in range(len(train_Images)):
    #Call function, passes needed variables, list of an image at a time, set in which to save in title, position, label number
    save_image(train_Images[i], 'train', (i+1), train_Labels[i])

#For loop until length of test_images(10,000)
for i in range(len(test_Images)):
    #Call function,  passes needed variables, list of an image at a time, set in which to save in title, position, label number
    save_image(test_Images[i], 'test', (i+1), test_Labels[i])
```
